```python
import matplotlib.pyplot as plt
```

## Plot a Point

Plot a dot at **x=5** and **y=10**

Remember that you have the `pyplot.scatter()` [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter.).

```python
plt.scatter(5,10)
```

## Plot Multiple Points

Plot several dots stored in a list of **dimensions (2, 13)** (meaning a list of lists)

```python
lists = [[21,14,16,2,6,23,7,8,9,3,1,12,14],[12,1,4,6,7,2,3,10,11,13,25,26,18]]
plt.scatter(lists[0],lists[1])
```

## Plotting a Line

Plot a line starting at **x=4, y=0** and ending at **x=20, y=4**. The function you should use is [pyplot.plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)

```python
plt.plot((4,20),(0,4))
```

## Plotting Multiple Curves

Plot 3 different curves on the same graph, still with pyplot.plot(). 

```python
years_x = [1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]
total_y = [1243, 1543, 1619, 1831, 1960, 2310, 2415, 2270, 1918]
years2_x = [1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010]
total2_y = [1240, 1540, 1610, 1830, 1960, 2315, 2410, 2275, 1910]
years3_x = [1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020, 2025]
total3_y = [1243, 1543, 1619, 1831, 1960, 2310, 2415, 2270, 1918]
plt.plot(years_x, total_y)
plt.plot(years2_x, total2_y)
plt.plot(years3_x, total3_y)
```
